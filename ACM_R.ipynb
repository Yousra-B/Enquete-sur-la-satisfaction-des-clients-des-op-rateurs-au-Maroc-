```r
library("ggplot2")
library("dplyr")
library("ggpubr")
library("FactoMineR") 
library("factoextra")
library("corrplot") 
```

# Internet

```r
Y=read.csv("C:/Users/USER/Desktop/data_1.csv",sep=";",dec=",")
Y
```

```r
dim(Y)
```

```r

a<-ggplot(data=Y,aes(x=service_wifi))+
geom_bar(fill="#FF9999", colour="black")
b<-ggplot(data=Y,aes(x=satisfaction_conx ))+
geom_bar(fill="#CC6666", colour="black")
c<-ggplot(data=Y,aes(x=freq_coupures.par.mois.))+
geom_bar(fill="#9999CC", colour="black")
d<-ggplot(data=Y,aes(x= change_cnx))+
geom_bar(fill="#66CC99", colour="black")
ggarrange(a,b,c,d + rremove("x.text"), 
          labels = c("A", "B", "C","D"),
          ncol = 2, nrow = 2)
```

```r
res.mca <- MCA (Y, graph = FALSE)
print(res.mca) 
```

```r
summary(res.mca)
```

```r
eig.val <- get_eigenvalue(res.mca) 
eig.val
```

```r
fviz_eig(res.mca, choice = "eigenvalue", 
               addlabels=TRUE)
```

```r
fviz_screeplot (res.mca, addlabels = TRUE, ylim = c (0, 60))
```

```r
fviz_mca_biplot (res.mca, repel = TRUE, 
 ggtheme = theme_minimal()) 
```

```r
var <- get_mca_var(res.mca) 
var 
```

```r
head(var$coord) 
```

```r
head(var$cos2) 
```

```r
head(var$contrib) 
```

```r
fviz_mca_var (res.mca, choice = "mca.cor", 
 repel = TRUE, 
 ggtheme = theme_minimal ()) 
```

```r
head(round(var$coord, 2), 4)
```

```r
fviz_mca_var (res.mca, 
 repel = TRUE, 
 ggtheme = theme_minimal ())
```

```r
head(var$cos2, 4) 
```

```r
fviz_mca_var(res.mca, col.var = "cos2", 
 gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), 
 repel = TRUE, 
 ggtheme = theme_minimal())
```

```r
corrplot(var$cos2, is.corr=FALSE)
```

```r
fviz_cos2(res.mca, choice = "var", axes = 1:2) 
```

```r
head(round(var$contrib,2), 4)
```

```r
fviz_contrib (res.mca, choice = "var", axes = 1, top = 15)
```

```r
fviz_contrib (res.mca, choice = "var", axes = 2, top = 15)
```

```r
fviz_mca_var(res.mca, col.var = "contrib", 
 gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), 
 repel = TRUE, 
 ggtheme = theme_minimal()
 )
```
