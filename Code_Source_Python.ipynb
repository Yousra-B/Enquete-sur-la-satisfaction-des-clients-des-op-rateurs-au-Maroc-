### Recodification de la variable prix_conx

```python
y = [str(data.prix_conx[i]) for i in range (0,len(data.prix_conx))]
data.prix_conx = [d.split("D")[0] for d in y]
data.prix_conx = [d.split("d")[0] for d in data.prix_conx]
data.prix_conx = [d.split("m")[0] for d in data.prix_conx]
data.prix_conx = [d.split(",")[0] for d in data.prix_conx]
data.prix_conx = [d.split()[0] for d in data.prix_conx]
```

### Recodification de la variable prix_mob

```python
data.prix_mob.value_counts()
```

```python
x = [str(data.prix_mob[i]) for i in range (0,len(data.prix_mob))]
data.prix_mob = [d.split("D")[0] for d in x]
data.prix_mob = [d.split("d")[0] for d in data.prix_mob]
data.prix_mob = [d.split("m")[0] for d in data.prix_mob]
data.prix_mob = [d.split(",")[0] for d in data.prix_mob]
data.prix_mob = [d.split()[0] for d in data.prix_mob]
```

```python
data.loc[data.prix_mob == "I"]= 0
data.loc[data.prix_mob == ".",'prix_mob']= 0
data.loc[data.prix_mob == "20/50",'prix_mob']= 35
data.loc[data.prix_mob == "Ça",'prix_mob']= 0
data.loc[data.prix_mob == "nan",'prix_mob']= 0

```

```python
data["prix_mob"] = [int(data.prix_mob[i]) for i in range(0,len(data.prix_mob))]
```

```python
data.prix_mob.value_counts()
```

```python
data.drop( data[ data["sexe"] == 0 ].index, inplace=True)
data.drop( data[ data["sexe"] == 250 ].index, inplace=True)
```

```python
def mod_usage(data,col):
    data[col] = [str(x) for x in data[col]]
    data[col] =data[col].map(lambda x:x.replace("Consultation Web","Consultation_Web"))
    data[col] =data[col].map(lambda x:x.replace("Envoi de fichiers","Envoi"))
    data[col] =data[col].map(lambda x:x.replace("Téléchargements de fichiers","Téléchargements"))
    data[col] =data[col].map(lambda x:x.replace("Réseaux sociaux : Facebook, Instagram, ....","Réseaux_sociaux"))
    data[col] =data[col].map(lambda x:x.replace("Streaming- Youtube, vidéos en ligne","Streaming"))
    data[col] =data[col].map(lambda x:x.replace("Chats et jeux en ligne","Chat&Jeux_en_ligne"))
    data[col] =data[col].map(lambda x:x.replace("Cours en ligne (Microsoft Teams, Zoom.m, Google Meet)","Cours"))
    data[col] =data[col].map(lambda x:x.replace("Les cours en ligne","Cours"))
    data[col] =data[col].map(lambda x:x.replace("Télévision","Television"))
    data[col] =data[col].map(lambda x:x.replace("Etude en ligne","Etude"))
    data.drop( data[ data[col] == "nan" ].index, inplace=True)
    data[col] =data[col].map(lambda x:x.replace(";"," "))
```

## Pie chart function

```python
def pie(data,col):
    labels = data[col].value_counts().keys().tolist()
    n = len(labels)
    if n==2:
        colors = ['#fb3999','#66b3ff']
    elif n==3:
        colors = ['#fb3999','#66b3ff', '#ffcc99']
    elif n==4:
        colors = ['#fb3999','#66b3ff', '#ffcc99',"#66f3ff"]
    elif n==5:
        colors = ['#fb3999','#66b3ff', '#ffcc99',"#66f3ff",'#adcc99']
    elif n==6:
        colors = ['#fb3999','#66b3ff', '#ffcc99',"#66f3ff",'#adcc99','#9a75c7']
    
    fig1, f1 = plt.subplots()
    f1.pie(data[col].value_counts(), labels=labels, colors = colors, autopct='%1.1f%%',shadow=False, startangle=60) 
    f1.axis('equal')
    plt.tight_layout()
    plt.show()
```
